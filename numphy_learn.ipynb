```python
# np.corrcoef()  gives a correlation matrix
# We want: correlation between hours studied and math score

corr_matrix = np.corrcoef(hours, math)
correlation = corr_matrix[0, 1]# top-right value in the 2x2 matrix

print(f'Correlation between Hours Studied and Math Score: {correlation:.4f}')
print()

if correlation > 0.5:
    print('Strong positive link - studying more = better math!')
elif correlation > 0.2:
    print('Weak positive link - slight trend, but not strong.')
else:
    print('Very weak/no link - hours studied alone may not predict score.')
```

```python
# Add calculated columns back to the dataframe
df['Total_Score']   = total_score
df['Average_Score'] = np.round(std_avg, 2)
df['Result']        = np.where(std_avg >= 50, 'Pass', 'Fail')

# Assign letter grades
def assign_grade(avg):
    if avg >= 90: return 'A+'
    elif avg >= 80: return 'A'
    elif avg >= 70: return 'B'
    elif avg >= 60: return 'C'
    elif avg >= 50: return 'D'
    else: return 'F'

df['Grade'] = [assign_grade(a) for a in std_avg]

def assign_division(avg):
    if avg >= 75: return 'Distinction'
    elif avg >= 60: return 'First Division'
    elif avg >= 45: return 'Second Division'
    elif avg >= 33: return 'Third Division'
    else: return 'Fail'

df['Division'] = [assign_division(d) for d in std_avg]

def calculate_gpa(avg):
    if avg >= 90:
        return 4.0  # Equivalent to an A
    elif avg >= 80:
        return 3.0  # Equivalent to a B
    elif avg >= 70:
        return 2.0  # Equivalent to a C
    elif avg >= 60:
        return 1.0  # Equivalent to a D
    else:
        return 0.0  # Equivalent to an F
    
df['GPA'] = [calculate_gpa(g) for g in std_avg]

# Save to a new CSV
df.to_csv('student_results.csv', index=False)

print(' Final results saved to student_results.csv!')
print()
print(' Grade Distribution:')
print(df['Grade'].value_counts().sort_index())
print()
print(' Division Distribution:')
print(df['Division'].value_counts().sort_index())
print()
print(' GPA Distribution:')
print(df['GPA'].value_counts().sort_index())
print()
print('Last 5 rows of final data:')
df.tail()
```
